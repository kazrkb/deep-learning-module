```python
# concatenation
import pandas as pd
data = {
    "x1": [0,0,1,1],
    "x2": [0,1,0,1],
    "y": [0,0,0,1]
}

AND = pd.DataFrame(data)
print(AND)
```

```python
X = AND.drop("y", axis=1)
y = AND["y"]
y.to_frame()
print(X)
print(y)
```

```python
bias = np.c_[X, -np.ones((len(X),1))]
print(bias)
```
